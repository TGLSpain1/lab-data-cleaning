## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd

weather_df = pd.read_csv("../weather-raw.csv")
```

```python
weather_df
```

```python
melted_weather_df = weather_df.melt(id_vars=["id", "year", "month", "element"], 
                                    var_name="day",
                                    value_name="temperature")



melted_weather_df["day"] = melted_weather_df["day"].str.replace("d", "").astype(int)

melted_weather_df
```

```python
null_cols = melted_weather_df.isnull().sum()
null_cols[null_cols > 0]
```

```python
max_df = melted_weather_df[melted_weather_df["element"] == "tmax"].sort_values(by=["month", "day"])
min_df = melted_weather_df[melted_weather_df["element"] == "tmin"].sort_values(by=["month", "day"])

merged_df = pd.merge(max_df, min_df, how = "left", on = ["month", "day"])

merged_df.drop(['id_y', 'year_y', "element_x", "element_y"], axis=1, inplace = True)

merged_df.rename(columns = {"temperature_x": "max_temperature", "temperature_y":"min_temperature"}, inplace = True)

merged_df.head(32)

```
